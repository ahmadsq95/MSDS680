# Bayes Theorem:
The probability of event A, given that event B has subsequently occurred can be defined as 

$$
P(A~|~B) = \frac{P(A) \cdot P(B~|~A)}{(~P(A) \cdot P(B~|A~)~) + (~P(A') \cdot P(B~|A'~)~)}
$$

**Note:** P(A')denotes P(not A) and the same applies for P(B').

```python
# loading required libraries
import pandas as pd 
import seaborn as sns
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer 
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics import confusion_matrix , accuracy_score , classification_report
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
from PIL import Image

import matplotlib.pyplot as plt
% matplotlib inline
```

    Downloading stop-words files from nltk package

```python
# nltk downloading stopwords files
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
spam = pd.read_csv('spam.csv', encoding='latin-1')
spam.head(10)
```

```python
spam.info()
```

    Dataset has 5 columns. We only need V1 and V2 all other columns are null and not useful. So, I dropped them and renamed V1 as Class, and V2 as text 

```python
#drop unused columns 
spam = spam.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
# rename columns
spam = spam.rename(columns={"v1":"class", "v2":"text"})
spam.head(10)
```

```python
# column that store text lentgh 
len_text=[]
for i in spam['text']:
    len_text.append(len(i))
spam['text_length']=len_text
```

```python
spam.head()
```

```python
plt.figure(figsize=(12,5))
spam[spam['class']=='spam']['text_length'].plot(bins=35,kind='hist',color='blue',label='spam',alpha=0.5)
plt.legend()
plt.xlabel('message length')
plt.show()
```

```python
plt.figure(figsize=(12,5))
spam[spam['class']=='ham']['text_length'].plot(bins=35,kind='hist',color='red',label='ham',alpha=0.5)
plt.legend()
plt.xlabel('message length')
plt.show()
```

    We can see that most of spam messages' length is between 100 and 150 while ham is between 40 and 70.

    The function below is to remove stop-words, punctuations, and cnvert upper-case t lower-case. 

```python
def process(sentence):
    text = re.sub('[^\w]+',' ',sentence).split()
    words = [x.lower() for x in text if x not in stopwords.words('english')]
    
    lemma = WordNetLemmatizer()
    word = [lemma.lemmatize(word,'v') for word in words ]
    word = ' '.join(word)
    return word
```

```python
spam['text'] = spam['text'].apply(process)
```

```python
spam.head()
```

    The data now is clean. Stopwords, punctuations removed, and upper-cases converted to lower-cases. 

```python
cv = CountVectorizer()
cv.fit(spam['text'])
print(cv.vocabulary_)
vector = cv.transform(spam['text'])
print(vector.shape)
print(type(vector))
print(vector.toarray())
```

```python
vectorizer = TfidfVectorizer()
vectorizer.fit(spam['text'])
print(vectorizer.vocabulary_)
print(vectorizer.idf_)
# encode document
vector = vectorizer.transform(spam['text'])
# summarize encoded vector
print(vector.shape)
print(vector.toarray())
```

    The Two code above are counting the words frequency in all documents (rows). they present same result, but I don't know how they work differently.

```python
text = spam.text[0]

# Create and generate a word cloud image:
wordcloud = WordCloud().generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

    The wordcloud above shows the most frequent word in the text. 

     Below is the modeling part. 
     
     we split the data into train and test wit ratio of 0.8 for train and 0.2 for test. 
     and we use Multinomial Naive Bayes model.

```python
# setting Target and features
y=spam['class']
x=spam['text']
```

```python
# splitting data
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=0,test_size=0.2)
```

```python
x_train=cv.fit_transform(x_train)
```

```python
nb=MultinomialNB()
nb.fit(x_train,y_train)
pred=nb.predict(cv.transform(x_test))
score=accuracy_score(y_test,pred)
score
```

    The model accuracy is 0.98 which is great. 
    Below is the confusion matrix where the true positive is 947, false positive is 2, true negative 12, and false negative is 154. 

```python
confusion_mt=confusion_matrix(y_test,pred)
```
